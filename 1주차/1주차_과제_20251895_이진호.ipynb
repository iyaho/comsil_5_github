# Problem 4
## NumPy의 np.dot()함수를 이용하여 주어진 하나의 matrix와 하나의 vector로 matrix-vector multiplication을 수행함.

```python
matrix_A = np.arange(32).reshape(8,4)
vector_x = np.arange(4)
matrix_A, vector_x
```

```python
mat_vec_mul_result = np.dot(matrix_A, vector_x)
print(mat_vec_mul_result)
```

NumPy의 matrix_A 행렬을 2차원 리스트로, 그리고 vector_x 벡터를 1차원 리스트로 변환

```python
list_2D_A = matrix_A.tolist()
list_1D_x = vector_x.tolist()
list_2D_A, list_1D_x
```

## 두 개의 리스트 list_2D_A와 list_1D_x를 이용하여 np.dot(matrix_A, vector_x) 함수와 같은 계산을 수행하는 mat_vec_mul_func() 함수를 작성하시오.

(참고: NumPy에서 제공하는 함수를 (ex, np.sum(), np.dot() 등) 이용하지 말고 반복문을 이용해서 코드를 작성할 것.)

```python
"""
Problem 4: 이 셀 안에 코드를 작성하시오.
"""
# 함수 작성
def mat_vec_mul_func(mat_A, vec_x):
  result_vec = [0] * len(mat_A)
  for i in range(len(mat_A)):
    for j in range(len(vec_x)):
      result_vec[i] += mat_A[i][j] * vec_x[j]
  return result_vec
```

```python
# 함수 호출
list_mat_vec_mul_result = mat_vec_mul_func(list_2D_A, list_1D_x)

# 결과 출력
print(list_mat_vec_mul_result)
```

# Problem 5
## NumPy의 np.dot()함수를 이용하여 주어진 두 개의 matrix로 matrix-matrix multiplication을 수행함.

```python
matrix_A = np.arange(32).reshape(8,4)
matrix_B = np.arange(64).reshape(4,16)
matrix_A, matrix_B
```

```python
mat_mat_mul_result = np.dot(matrix_A, matrix_B)
print(mat_mat_mul_result)
```

NumPy의 matrix_A 와 matrix_B 행렬을 2차원 리스트로 변환

```python
list_2D_A = matrix_A.tolist()
list_2D_B = matrix_B.tolist()
list_2D_A, list_2D_B
```

## 두 개의 리스트 list_2D_A와 list_2D_B를 이용하여 np.dot(matrix_A, matrix_B) 함수와 같은 계산을 수행하는 mat_mat_mul_func() 함수를 작성하시오.

(참고: NumPy에서 제공하는 함수를 (ex, np.sum(), np.dot() 등) 이용하지 말고 반복문을 이용해서 코드를 작성할 것.)

```python
"""
Problem 5: 이 셀 안에 코드를 작성하시오.
"""
# 함수 작성
def mat_mat_mul_func(mat_A, mat_B):
  result_mat = []
  for li in mat_A:
    a = []
    for j in range(len(mat_B[0])):
      s = 0
      for i in range(len(mat_B)):
        s += li[i] * mat_B[i][j]
      a.append(s)
    result_mat.append(a)
  return result_mat
```

위에 구현한 row_sum() 함수를 호출하여 반환된 결과값을 출력함.

```python
# 함수 호출
list_mat_mat_mul_result = mat_mat_mul_func(list_2D_A, list_2D_B)

# 결과 출력
print(list_mat_mat_mul_result)
```

```python
vector_x = np.arange(100000)
vector_y = np.arange(100000)
```

```python
def vec_vec_mul_func(vec_x, vec_y):
    result_vec = 0
    for i in range(len(vec_x)):
      result_vec += vec_x[i] * vec_y[i]
    return result_vec
```

```python
start = time.time()
np.dot(vector_x, vector_y)
end = time.time()
print(f"NumPy함수 : {end - start:.5}초")
nump = end - start
```

따라서 결론은 연산하는 배열의 크기가 커질수록 NumPy 내장 함수가 더 효율적이다는 것이다.
